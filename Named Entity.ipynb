```python
nlp = spacy.load("en_core_web_sm") # pos
nlp2 = spacy.load('en_core_web_md') # similarity
```

```python
grammar = r"NP: {<DT|CD|NN.*|JJ*>+<,|CC|JJ|NNP+>{1,}}"
# grammar = r"NP: {<|DT|CD|NN.*|JJ*>+<,|NN|CC|JJ|NNP.+>{1,}}"
cp = nltk.RegexpParser(grammar)
STOPWORDS = ['account', 'pay', 'statement', 'date', 'statement', '80 Lamberton Road Windsor'.lower(),
             'balance', 'fee', 'limited', 'net', 'value', 'ss & c', 'avenue', '%', 'cc', 'for', '|', 
            'Custom House Fund'.lower(), 'investor', 'service', 'street', 'issue', 'new york', 
            'Chicago , IL'.lower(), 'ny', 'Michael Leffell'.lower()]

name_entities = {}
temp = {}

for file in files:
    # initialize 
    s1 = []
    new_words_list = []
    s3 = []
    names = []
   
    fileID = file.split('/')[1].split('.')[0]
    
    with open(file, 'rb') as f:
        pages = pdftotext.PDF(f)
    text = ''
    for page in pages:
        text += page
        
    sentences = text.split('\n')
#     print(sentences)
    for sentence in sentences:
        if sentence.split(':') != []:
            s2 = sentence.split('  ')
        else:
            s2 = sentence.split(':')
        s2 = [s.strip() for s in s2 if s]
        if len(s2)==1 and len(s2[0].split(' '))>15:
    #         print(s2)
            continue
        else:
            s1.extend(s2)
#     print(s1)
#     s3 = [phrase for phrase in s1 if not re.search('\d{2,}', phrase) and not re.search(':+', phrase) and not re.search('/+', phrase)]

    spacy_filtered = []
    for s in s1:
        ents = [(e.text, e.start_char, e.end_char, e.label_) for e in nlp(s).ents]
#         if len(ents)==1 and ents[0][-1] == 'PERSON':
#             continue
        if 'LP' in s or 'LLC' in s:
            spacy_filtered.append(s)
        elif ents != []:
            spacy_filtered.append(s)
#     print(spacy_filtered)
    words_list = [nltk.word_tokenize(sentence) for sentence in spacy_filtered]
#     print(words_list)
#     for words in words_list:
#         new_words_list.append([word for word in words if word.lower() not in STOPWORDS])
    
#     for item in new_words_list:
#         s3.append([word for word in item if not re.search('\d+', word) and not re.search(':+', word) and not re.search('/+', word)])  
#     s3 = [item for item in s3 if item]
    
    tagged = [nltk.pos_tag(word) for word in words_list]
    tagged = [item for item in tagged if item != []]
#     print(tagged)
    for elm in tagged:
        entity = cp.parse(elm)
        for tree in entity.subtrees():
            ne = ''
            if tree.label() == 'NP':
        #         print(tree.leaves())
                for leaf in tree.leaves():
                    ne += leaf[0]
                    ne += ' '
#             print(ne)
#             if ne and re.search('[A-Z]+', ne):
            if ne and (ne.strip()[0].isupper() or ne.strip()[0].isdigit()):
                if len(ne.split('-')) > 1:
                    for phrase in ne.split('-'):
                        if phrase:
                            safe = True
                            for word in STOPWORDS:
                                if word not in ne.lower():
                                    continue
                                else:
                                    safe = False
                            if safe:
                                doc = nlp(ne)
                                for ent in doc.ents:
                                    if ent.label_ == 'DATE':
                                        ne = ne.replace(ent.text, '')
#                                 names.append(ne.strip())
                                doc2 = nlp(ne)
                                if len(doc2.ents)==1 and doc2.ents[0].text.isdigit():
                                    pass
                                elif len(ne)>3:
                                    names.append(ne.strip())
                else:
                    safe = True
                    for word in STOPWORDS:
                        if word not in ne.lower():
                            continue
                        else:
                            safe = False
                    if safe:
                        doc = nlp(ne)
                        for ent in doc.ents:
                            if ent.label_ == 'DATE':
                                ne = ne.replace(ent.text, '')
                        doc2 = nlp(ne)
#                         print(doc2.ents)
                        if len(doc2.ents)==1 and doc2.ents[0].text.isdigit():
                            pass
                        elif len(ne)>3:
                            names.append(ne.strip())
                    
                            
                        
    temp[fileID] = names
#     print(names)                
    if len(set(names)) > 2:
        list_ = []
        for p1 in set(names):
            for p2 in set(names):
                if p1!=p2:
                    if len(nlp(p1).ents)==1 and len(nlp(p2).ents)==1:
                        if not nlp(p1).ents[0].label_==nlp(p2).ents[0].label_=='PERSON':
                            list_.append((p1, p2, nlp2(p1).similarity(nlp2(p2))))
                    else:
                        list_.append((p1, p2, nlp2(p1).similarity(nlp2(p2))))
        final_names = (sorted(list_, key=lambda x: x[-1], reverse=True)[0][0], sorted(list_, key=lambda x: x[-1], reverse=True)[0][1])
        
        name_entities[fileID] = list(final_names)
    else:
        name_entities[fileID] = list(set(names))
```

```python
entity_and_fundName = {}
for file in name_entities:
    if len(name_entities[file]) == 2:
        if 'fund' in name_entities[file][0].lower():
            entity_and_fundName[file] = {'fundName': name_entities[file][0], 'entity': name_entities[file][1]}
            continue
        elif 'fund' in name_entities[file][1].lower():
            entity_and_fundName[file] = {'fundName': name_entities[file][1], 'entity': name_entities[file][0]}
            continue
    
        urls1 = []
        urls2 = []
        try:
            for url in search(name_entities[file][0], stop=10):
                urls1.append(url)
            for url in urls1:
                if 'whalewisdom' in url:
                    entity_and_fundName[file] = {'fundName': name_entities[file][0], 'entity': name_entities[file][1]}
                    break
        except:
            pass

        if not entity_and_fundName.get(file):
            try:
                for url in search(name_entities[file][1], stop=10):
                    urls2.append(url)
                for url in urls2:
                    if 'whalewisdom' in url:
                        entity_and_fundName[file] = {'fundName': name_entities[file][1], 'entity': name_entities[file][0]}
                        break
            except:
                pass
    
    else:
        if 'fund' in name_entities[file][0].lower():
            entity_and_fundName[file] = {'fundName': name_entities[file][0]}
            continue
    
        urls1 = []
        urls2 = []
        try:
            for url in search(name_entities[file][0], stop=10):
                urls1.append(url)
            for url in urls1:
                if 'whalewisdom' in url:
                    entity_and_fundName[file] = {'fundName': name_entities[file][0]}
                    break
        except:
            pass
```
