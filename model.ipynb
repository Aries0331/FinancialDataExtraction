```python
def category_to_numeric(x):
    if x == 'Account Statement':
        return 1
    elif x == 'Distribution Notice':
        return 2
    elif x == 'Call Notice':
        return 3
    else:
        return -1
```

```python
def convert_to_text(path=None):
    row = 0
    for file in glob.glob('docs/raw_data/*.pdf'):
        content = []
        with open(file, 'rb') as f:
            pdf = pdftotext.PDF(f)
    #         print (pdf)
            for page in pdf:
                page = page.replace('\n', ' ')
    #             print(page)
                content += nltk.sent_tokenize(page)
    #     print(content)
        new_df['text'][row] = content[0]
        row += 1
```
