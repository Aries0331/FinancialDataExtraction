```python
import spacy
nlp = spacy.load('en')
def compare(p1):
    sims = []
    p1 = nlp(p1)
    columns = ["Beginning Balance", "Beginning Date", "Capital Commitment", "Capital Contributions", 
          "Total Distributions", "Capital Distributions", "Ending Balance", "Data Date", "Entity", 
          "Investment Name", "Outstanding", "Total Paid in Capital", "MTD Return %", "MTD Return $",
          "Share Class", "Unfunded Commitment", "Withdrawal"]
    score = 0
    most_sim = ''
    for word in columns:
        temp = p1.similarity(nlp(word.lower()))
        if temp > score:
            score = temp
            most_sim = word
        sims.append(temp)
#     print(sims)
    return most_sim
```

```python
compare('balance at beginnine')
```
